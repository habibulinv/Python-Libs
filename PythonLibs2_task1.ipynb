### Вопрос 1

Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

### Ответ

**Precision**

1) Пример №1.
Боремся с оттоком покупателей в торговой сети. Хотим предсказать клиентов, которые могут от нас уйти и дать им глубокую скидку.
Если дать такую скидку большому кол-во покупателей, можно сжечь весь бюджет. По этом лучше кому-то 
не дать, чем дать скидку тем кто в ней не нуждался и на кото она не сработает.


2) Пример №2
Камеры контроля скорости на базе машинного зрения. Задача выявление нарушений и выписка штрафов. 
Очень важно не ошибиться и отправить штраф только тем кто с максимальной вероятностью наршул ПДД. Лучше пропустить некоторых нарушителей, само наличие камеры в целом будет способствовать уменьшению нарушений.


**Recall**

1) Пример №1.
Нужно увеличить продажи товара в новой категории. Мы решаем выбрать сегмент покупателей на основе их истории
покупок для того чтобы отправить им уникальные спец предложения из этой категории.
Тут нам больше важен охват и мы будем максимизировать Recall, так как если мы промахнемся рисков почти нет. 


2) Пример №2
Полиция обьявила план перехват и надо найти по камерам автомобиль отвечающий описанию свидетелей. Система видеораспознавания расположенная перед поставми ГИБДД должна выявлять как можно больше кол-во автомобилей отвечающих описанию, чтобы сотрудники органов вовремя могли остановить автомобиль и проверить.



### Вопрос 2

Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

### Ответ

Если мы хотим получить взвешанную оценку нашему предсказанию, то относительные показатели не стоит
просто усреднять. Precision - это про точность, а Recall - про полноту.
Нам нужно найти не просто среднее между ними, а найти оценку, учитываю вклад каждой величины.
И для этого применение средней гармонической подходит лучше всего.
