### Вопрос 1

В чём различие между зависимыми и независимыми выборками?

### Ответ

В зависимых выборках есть какой-либо критерий, который связывает объет из одной выборки с объектом из другой.
Самый популярный пример зависимой выборки, это "до-после". Т.е. выборки где одни и те же объекты представлены до и после некого воздействия. Или, например, продажи определенной категории товаров в одной торговой сети и в другой.
В независимых выборках у объектов нет явной связи или соответствия. 

### Вопрос 2

Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

### Ответ

Параметрические статистические критерии применяються только при нормальном распределении и только при достаточно большом объеме данных.
Непараметрические применяются когда распределение неизвестно. Когда мало объектов. И я так понял, чаще всего при исследовании категориальных признаков.
