```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
%config InlineBackend.figure_format = 'svg'
```

```python
DATASET_PATH = './creditcard.csv'
PREPARED_DATASET_PATH = './creditcard.csv'
```

```python
dc = pd.read_csv(DATASET_PATH, sep=',')
dc.head(10)
```

```python
dc['Class'].value_counts()
```

```python
dcplt = dc['Class'].value_counts()
dcplt.plot(kind='bar')
plt.yscale(value="log")
plt.show()
```

```python
dc[dc['Class']<1]['V1']
```

```python
cls0 = dc[dc['Class']==0]['V1']
cls1 = dc[dc['Class']==1]['V1']
```

```python
cls0.hist(bins=20, orientation="vertical", color="Gray", alpha=0.5, density=True)
cls1.hist(bins=20, orientation="vertical", color="Red", alpha=0.5, density=True)
plt.xlabel("Class")
plt.legend(labels=["Class 0", "Class 1"])
plt.show()
```
